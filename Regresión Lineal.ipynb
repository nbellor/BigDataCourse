## Introducción

El análisis de regresión es una técnica estadística para investigar y modelar la relación entre variables. Las aplicaciones de la regresión son numerosas y ocurren en casi todos los campos, incluyendo la ingeniería, física, química, economía, administración, biología y ciencias de la vida y sociales. ## Historia

La primera forma de regresión fue el método de mínimos cuadrados, el cual fue publicado por Legendre en 1805, y por Gauss en 1809. Tanto Legendre como Gauss aplicaron el método al problema de determinar, a partir de mediciones astronómicas, la orbita de los cuerpos al rededor del sol. Gauss publicó un avance adicional a la teoría de mínimos cuadrados, incluyendo una versión del teorema de Gauss-Markov.

El término 'regresión' fue acuñado por Francis Galton en el siglo 19 para describir un fenómeno biológico. El fenómeno consistía en que las alturas de los descendientes de ancestros altos tendían a 'regresar' hacia un promedio normal. Para Galton, la regresión sólo tenía este significado biológico, pero su trabajo fue posteriormente extendido por Udny Yule y Karl Pearson a un contexto estadístico más general.

Los métodos de regresión continúan siendo un área de investigación activa. En décadas más recientes nuevos métodos han sido desarrollados para regresión robusta, regresión con respuestas correlacionadas como las series temporales o las curvas de crecimiento, regresión en la cual las variables predictoras o respuesta son curvas, imágenes, u otros objetos complejos. [2] 

## Planteamiento

Se considera el modelo $$y=\alpha + \beta x $$ el cual describe una recta con pendiente $\beta$ e intercepto en el eje $y$, $\alpha$. En lo general dicha relación entre variables no se mantiene exactamente y a las desviaciones las llamamos errores. Supongamos que se tienen $n$ observaciones. Podemos describir la relación entre $x$ y $y$, involucrando los errores $\varepsilon_i$ de la siguiente forma $$y_i = \alpha + \beta x_i + \varepsilon_i.$$ El objetivo es estimar valores $\hat \alpha$ y $\hat \beta$ que sean los 'mejores' en algún sentido para el conjunto de observaciones. En esta ocasión buscamos minimizar la suma de los cuadrados de los residuales $\hat \varepsilon_i = y_i - \alpha - \beta x_i$ , esto es: $$\min_{\alpha, \beta} \sum_{i=1}^{n} \hat \varepsilon_i $$

## Ejemplo

Para el ejemplo vamos a usar el conjunto de datos Iris. Tomaremos las variables largo y ancho del sépalo para los individuos de la variedad Setosa. Primero cargamos los datos desde un archivo .csv on ayuda del paquete 'pandas'.

```python
import numpy as np
import pandas as pd

iris = pd.read_csv('iris.csv')
```

Ahora podemos realizar el gráfico de dispersión con 'matplotlib'.

```python
import matplotlib.pyplot as plt

plt.plot(iris['sepal.length'][iris.variety == 'Setosa'], iris['sepal.width'][iris.variety == 'Setosa'], '.')
plt.xlabel('largo del sépalo (cm)')
plt.ylabel('ancho del sépalo (cm)')
plt.title('Iris Setosa')
plt.show()
```

En el paquete 'sklearn' podemos encontrar la clase 'LinearRegression' para llevar a cabo la estimación de los parámetros.

```python
from sklearn.linear_model import LinearRegression as lm

X = iris.iloc[:50, 0].values.reshape(-1, 1)
Y = iris.iloc[:50, 1].values.reshape(-1, 1)
lm1 = lm()
lm1.fit(X, Y)
print('pendiente estimada ', lm1.coef_[0][0])
print('intercepto estimado ', lm1.intercept_[0])
```

Entonces podemos interpretar la pendiente como que por cada centímetrro adicional en el largo del sépalo, el ancho del mismo tendrá 0.8 centímetros adicionales. Por último, podemos visualizar la recta estimada con la función 'predict'.

```python
Y_pred = lm1.predict(X)
plt.plot(X, Y, '.')
plt.plot(X, Y_pred, color='red')
plt.show()
```
